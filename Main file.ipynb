## Cleaning

Checking the number of null and non-null values before dataset cleaning.

```python
df.isnull().sum()
```

```python
df.info()
```

Assigning more user-friendly column names.

```python
df = df.rename(columns={
    'current/discounted_price': 'discounted_price',
    'bought_in_last_month': 'monthly_sales'
})
```

Removing values that are not relevant or useful for further analysis and modeling.


```python
df.drop(['price_on_variant','buy_box_availability',	'sustainability_badges', 'image_url',	'product_url', 'collected_at'], axis=1,inplace=True)
```

identification and removing rows without prices

```python
rows_without_any_price = (df['listed_price'] == 'No Discount') & (df['discounted_price'].isna())
df = df[~rows_without_any_price]
```

For rows with no discount, 'listed_price' just says "No Discount" instead of holding a price.
Since there's no discount, the listed price equals the discounted price.

```python
df.loc[df['listed_price'] == 'No Discount', 'listed_price'] = df.loc[df['listed_price'] == 'No Discount', 'discounted_price']
```

processing numeric columns, assigning them to other columns, and changing data types

```python
#Removing $ symbol from the beginning of price value
df['listed_price'] = df['listed_price'].str.replace(r'^[^\d]+', '', regex=True)

df.loc[df['discounted_price'].isna(), 'discounted_price'] = df.loc[df['discounted_price'].isna(),'listed_price']
```

```python
numeric_cols = ['listed_price', 'discounted_price', 'number_of_reviews']
for col in numeric_cols:
    df[col] = df[col].str.replace(',', '', regex=False).astype(float)
```

For products with no reviews and no monthly sales, I assumed these values were equal to 0.

However, for missing ratings, I used the average rating to avoid artificially lowering the overall score.

For missing delivery details, I assumed that the data was simply unavailable.

```python
df['number_of_reviews'] = df['number_of_reviews'].fillna(0)
df['monthly_sales'] = df['monthly_sales'].replace('', np.nan).fillna(0)

df['rating'] = df['rating'].apply(lambda x : str(x).split(' ')[0])
df['rating'] = df['rating'].astype(float)

df['rating'] = df['rating'].fillna(round(df["rating"].mean(),1))

df['delivery_details'] = df['delivery_details'].fillna('No data')
```

Identifying value errors in the monthly_sales column and removing them.

&

Extracting proper float values from the monthly_sales Column

```python
df['monthly_sales'] = df['monthly_sales'].str.replace(r'[^0-9K]+$', '', regex=True)

df['monthly_sales'].value_counts().tail(10)

df = df[df['monthly_sales'].astype(str).str.match(r'^[0-9K]+$')]

mask = df['monthly_sales'].str[-1] == 'K'
df.loc[mask, 'monthly_sales'] = df.loc[mask, 'monthly_sales'].apply(lambda x: np.double(str(x)[:-1]) * 1000)

df['monthly_sales'] = df['monthly_sales'].astype(float)
```

Identifying error values in the delivery_details column and removing them

```python
df['delivery_details'].value_counts().head(50)
```

Cutting out the paid delivery date and setting NaN for fields without a specified delivery date

```python
df['delivery_details'] = df['delivery_details'].astype(str).str.extract(r'([A-Z][a-z]{2}(?:[-\s]*\d+)*)$')
#it extracting for example "Jan 15" from "Free delivery Jan 15"

df.loc[df['delivery_details'].isna(), 'delivery_details'] = "No data"
```

```python
df['delivery_details'].head(50)
```

Creating a new column to store dates for easier access and analysis

```python
df['delivery_date'] = pd.to_datetime(
    df['delivery_details'] + ' 2025',
    format='%b %d %Y',
    errors='coerce'
)
```

```python
missing_date_mask = df['delivery_date'].isna()
df.loc[missing_date_mask, 'delivery_date'] = pd.to_datetime(
    df.loc[missing_date_mask, 'delivery_details']
      .str.extract(r'([A-Z][a-z]{2}\s*\d{1,2})')[0]
      .add(' 2025'),
    format='%b %d %Y',
    errors='coerce'
)

mask = df['delivery_details'] == "No data"
df.loc[mask,'delivery_date'] = np.nan
```

removing useless column

```python
df.drop(columns='delivery_details', inplace=True)
```

changing sponsored column type to True/False

```python
df['is_sponsored'].value_counts()
```

```python
df['is_sponsored'] = df['is_sponsored'] != 'Organic'
```

Identifying issues in the is-couponed column

```python
df['is_couponed'].value_counts()
```

Creating a helper function to set missing coupon values to 0 and split percentage and raw price coupons

creating new column and removing old one

```python
def compute_coupon_profit(row):
    coupon = row['is_couponed']
    if coupon in ('No Coupon', 'No'):
      return 0.0
    coupon = coupon.split()[1]
    if coupon.endswith('%'):
        percent = float(coupon.rstrip('%'))
        return round(row['discounted_price'] * percent / 100.0, 2)
    return float(coupon.replace('$', ''))

df['coupon_profit'] = df.apply(compute_coupon_profit, axis=1)

df.drop(columns='is_couponed', inplace=True)
```

Set "is_best_seller" to boolean type column.

```python
df['is_best_seller'] = df['is_best_seller'].isin(["Best Seller", "Amazon's"])
df['is_best_seller'] = df['is_best_seller'].astype(bool)
df['is_best_seller'].value_counts()
```

```python
before = len(df)
df = df.drop_duplicates()
print(f"Deleted {before - len(df)} duplicates {len(df)}  rows remains")
```

Creating new column with montly revenue for further analysis

```python
df['last_month_revenue'] = df['monthly_sales'] * df['discounted_price']
```

Dataset is clean

### Dataset Analysis

The distribution is left-skewed, which is common for e-commerce platforms

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="rating", color="steelblue")

plt.title("Distribution of product ratings")
plt.xlabel("Rating")
plt.ylabel("Number of products")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("charts/rating.png", dpi=150, bbox_inches="tight")
plt.show()

```

As we can see, the rating level is correlated with the number of orders for a given product. Products with a record number of copies sold have no less than 4.5 stars.

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(
    data=df,
    x="rating",
    y="monthly_sales"
)

plt.title("Correlation: number of copies sold and rating")
plt.xlabel("Rating")
plt.ylabel("Monthly sales")
plt.tight_layout()
plt.savefig("charts/num_copies_sold_monthly_sales.png", dpi=150, bbox_inches="tight")
plt.show()
```

I decided to apply a logarithmic function to better illustrate the distribution of the data on the chart. However, we can see that the discount value is not correlated with the number of copies sold.

```python
df['discount_amount'] = df['listed_price'] - df['discounted_price']
```

```python
df["discount_amount_log"] = np.log(df["discount_amount"] + 1)
```

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(
    data=df,
    x="discount_amount_log",
    y="monthly_sales",
)

plt.title("Correlation: number of copies sold and discount amount")
plt.xlabel("Discount Amount log")
plt.ylabel("Monthly Sales")
plt.tight_layout()
plt.show()
```

To better assess the relationships, I used a correlation matrix.
It confirms no correlation between the discount and number of sold copies.

Shows that the link between product ratings and number of sold copies is weaker than I initially assumed.

We can also see that sponsorship affects sales, but not in a very significant way.

```python
corr = df[['monthly_sales', 'discount_amount', 'rating', 'is_sponsored']].corr()

sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdBu_r")

plt.title("Correlation matrix for the number of copies sold, product rating, discount amount, and sponsorship")
plt.tight_layout()
plt.savefig("charts/copies_corr.png", dpi=150, bbox_inches="tight")
plt.show()
```
